- Using a dict() or Typecasting

Syntax:
     
     dict(iterable)

```python
t = (('Name', 'Atul'), ('Age', 63)) # tuple of a tuple
print(t, type(t))
```

### Case - 3

#### Zip ()

    - It takes two iterable as an argument
    - It maps all the elements of fisrt iterable to all the element of the second iteable one by one in a sequnece
    - It fetch the values we can pass the object to list [], tuple () or a dictionary {}
    
__Note__:-

- While creating a dictionary using zip(), we should remember that it should have two iterable object as an argument
- The number of elements inside the iterable should be the same
- If the number of elements are not equal it will map only the elements which have respectively leaving the element as it is in the one having greater number of elements

a = 'abcde'

lst = [10,20,30,40,50]

                        a  b  c  d  e
                        1  2  3  4  5
                       10 20 30 40 50

```python
a = 'abcde'
lst = [10,20,30,40,50]

z = zip(a, lst) #here the values in a will be considered as keys and the values in the lst will be considered as the values
print(z)
```

```python
print(dict(z))
```

```python
# Wehave create two empty list that will store the keys and the values associated to it
keys = []
values = []

# taking input from the user
for i in range(4):
    a = eval(input('Enter a key:'))
    keys.append(a)

    b = eval(input('Enter a value'))
    values.append(b)

#zipping the two list (iterable) together
z = zip(keys, values)
print(dict(z))
```

### Case - 5

- Using Dictionary comprehension

```python
# create a dictionary takeing numbers as the keys and their squares as the values
d = {i:i**2 for i in range(1,11)}
print(d)
```

```python
# create a dictionary takeing even numbers as the keys and their squares as the values
d = {i:i**2 for i in range(1,11) if i%2 == 0}
print(d)
```

```python
# Create a dictionary where the letter A to Z are the keys and their associated ASCII values are hoing to be the values
ascii_dict = {letter: ord(letter) for letter in 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'}

print(ascii_dict)
```

```python
d={chr(i):i for i in range(ord("A"),ord("Z")+1)}
print(d)
```

```python
# Wehave create two empty list that will store the keys and the values associated to it
keys = []
values = []

# taking input from the user
for i in range(4):
    a = input('Enter a key:')
    keys.append(a)

    b = eval(input('Enter a value: '))
    values.append(b)

#zipping the two list (iterable) together
z = dict(zip(keys, values))
print(z)
```

### Nested Dictionary

- A dictionary within another dictionary is known as nested dictionary
- Where a dictionary is mapped to a key of another dictionary then it is known as Nested dictionary

Syntax:

    {key:{key:value}}

```python
emp_1 = {'name':'Rajesh', 'age':32, 'ID':101, 'salary':45000}
emp_2 = {'name':'Pooja', 'age':28, 'ID':102, 'salary':48000}
emp_3 = {'name':'Ali', 'age':29, 'ID':103, 'salary':45000}
```

```python
emp_detail = {'emp_1_detail':emp_1, 'emp_2_detail':emp_2, 'emp_3_detail':emp_3}
```

```python
emp_detail
```

```python
emp_detail['emp_1_detail']
```

```python
emp_detail['emp_2_detail']
```

```python
emp_detail['emp_1_detail']['name']
```

```python
emp_detail['emp_3_detail']['age']
```

# Dictionary are mutable

```python
emp_detail['emp_1_detail']['salary'] = 75000
```

```python
emp_detail
```

```python
emp_detail['emp_1_detail'] = 'Name'
```

```python
emp_detail
```

### How to iterate over a nested dictionary

```python
emp_1 = {'name':'Rajesh', 'age':32, 'ID':101, 'salary':45000}
emp_2 = {'name':'Pooja', 'age':28, 'ID':102, 'salary':48000}
emp_3 = {'name':'Ali', 'age':29, 'ID':103, 'salary':45000}
```

```python
emp_detail = {'emp_1_detail':emp_1, 'emp_2_detail':emp_2, 'emp_3_detail':emp_3}
```

```python
emp_detail
```

```python
# here we are iterating over the keys of emp_detail
for i in emp_detail:
    print(i)
```

```python
# we want to iterate over the values

for i in emp_detail:
    print(emp_detail[i])
```

```python
#Fetch out the names of all the employees from the nested dictionary

for i in emp_detail:
    print(emp_detail[i]['name'])
```

```python
for i in emp_detail:
    print(emp_detail[i]['age'])
```

```python
# fetch out the employee name along with salary and age

for i in emp_detail:
    print(emp_detail[i]['name'], emp_detail[i]['age'], emp_detail[i]['salary'])
```

```python
emp_1 = {'name':'Rajesh', 'age':32, 'ID':101, 'salary':45000, 'city':'Cochin'}
emp_2 = {'name':'Pooja', 'age':28, 'ID':102, 'salary':48000, 'city':'Goa'}
emp_3 = {'name':'Ali', 'age':29, 'ID':103, 'salary':45000,'city':'Jaipur'}
```

```python
emp_detail = {'emp_1_detail':emp_1, 'emp_2_detail':emp_2, 'emp_3_detail':emp_3}
```

```python
emp_detail
```

```python
# change the city of rajesh

for i in emp_detail:
    if emp_detail[i]['name'] == 'Rajesh':
        emp_detail[i]['city'] = 'Mumbai'
```

```python
emp_detail
```

### keys()

- It returns a dictionary key object, the dictionary key object contains a list of keys from the dictionary

```python
d = {i:i**2 for i in range(1,11)}
print(d)
```

```python
a = d.keys()
print(a, type(a))
```

```python
emp_detail.keys()
```

#### Inner Keys

```python
for i in emp_detail:
    print(emp_detail[i].keys())
```

### values()

- It returns a dictionary value object, the dictionay values object contains a list of the values from the dictionary

```python
d.values()
```

```python
emp_detail.values()
```

#### Inner values

```python
for i in emp_detail:
    print(emp_detail[i].values())
```

### items()

- It returns a dict item object, that contains a list of tuple where every tuple has two values
- The first element of a tuple is the key and the second element of the tuple is the values associated to that particular key

```python
d.items()
```

```python
emp_detail.items()
```

#### Inner Items

```python
for i in emp_detail:
    print(emp_detail[i].items())
```

```python
emp_detail
```

```python
for i, j in emp_detail.items():
    print(i)
    print(j)
```

```python
for i, j in emp_detail.items():
    print(i) #fetches the keys
    print(j) # fetches the values
    print('----------------------------------')
    for x in j.items(): # I am again using item method to get the value of the inner dictionary
        print(x)
    print()
```
