#### Identity Operator

- it is used to check whether two values are having the same memory location (or address) or not
- it compares the memory address of the value, they donot compare the values
- it also returns boolean value

    - is
    - is not

```python
a = 10
b = 10

print(a is b)
```

```python
a = 10.0
b = 10.0

print(a is b)
```

```python
print('10.0' is '10.0')
```

```python
[1] is [1]
```

```python
a = [1]
b = [1]

print(id(a), id(b))
```

### Bitwise left shift operator (<<)

```python
8 bit system
16 bit system
32 bit system
```

```python
32 bit system ------ it will take 32 bits to represent a value
```

#### Steps for the Bitwise Left shift Operator :-
    
    - Convert the number into the binary form
    - express the number in the 32 bit system format
    - we have to remove the specified numbers of bits from the left and going to add the same number of 0's in the right
    - convert the result back to my decimal equivalent

```python
bin(10)
```

```python
00000000000000000000000000001010 #this is number 10 represented in the 32 bit system format
```

```python
45<<4
```

```python
bin(45)
```

```python
26<<6
```
