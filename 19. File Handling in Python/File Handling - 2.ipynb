```python
# Permissions that we give while opening the file. 

# r for read, t for text 

# Operations permitted  Text Mode  Binary Mode  Handle Position     Notes

# read only             'r'        'rb'         Beginning of file   Default mode of file opening. If file does not exist, 
#                                                                   raises an I/O error.
# read and write        'r+'       'rb+'        Beginning of file   If file does not exist, raises an I/O error. By default,
#                                                                   any write operations are performed at beginning of file.
# write only            'w'        'wb'         Beginning of file   If file exists, truncates file(deletes all contents)
#                                                                   else creates a file for writing into
# write and read        'w+'       'wb+'        Beginning of file   If file exists, truncates file(deletes all contents)
#                                                                   else creates a file for writing into
# append                'a'        'ab'         End of file         If file does not exist, creates file. If file exists, 
#                                                                   new data will be written at end of file. 
# append and read       'a+'       'ab+'        End of file         If file does not exist, creates file. If file exists, 
#                                                                   new data will be written at end of file

# Exclusive creation     'x'        'xb'        Beginning of file   Opens a file for exclusive creation(new file). If 
#                                                                   the file already exists raises an error.
# Text mode              't'                                        Default mode. You could say, a file without mode
#                                                                   specified opens a file in Text mode for reading (both
#                                                                   are defaults)

# Please remember that handle(cursor) position mentioned above are Default positions. However, once operations have been 
# performed and cursor has been moved - write, read operations are performed from the current cursor position. 
```

```python
# What is the difference between r+ and w+ i.e. read and write, write and read?

# r+ - my file does not get truncated - content remains. THEN I can read from it, send cursor anywhere I want(if I want) and
# then write wherever I want. 

# w+ - my file gets truncated - so I must write something into it, send cursor wherever I want and then read from there. 


# r+ - read and write.... why dont we have read and append?
# w+ - write and read .... why dont we have write and append?
# a+ - append and read .... why dont we have append and write?

# append does not truncate file
# append sends handle to end (which can always be changed by using f.seek())

# Otherwise everything else same as write.
```

#### Other parameters we can use with open function.

```python
# The open function takes several parameters. Some of these are:

#1. File name (including path if required) as string - mandatory
#2. Mode - optional. If not specified opens as a text file in read only mode. 
#3. Buffer size - optional - Default is system dependent (usually 4096 or 8192 bytes). 0 means no buffering, 1 means until
# new line. Otherwise buffersize can be specified. 
#4. Encoding - Optional Specifies encoding (Unicode - <UTF8,UTF16, UTF32>, ascii, cp1252 etc.)
#5. Errors - Optional - 
#   'strict’ (Default) - raises an UnicodeError exception on failure
#   ‘backslashreplace’ - the unencodable character is replaced by a backslash and hexadecimal values of the char
#   ‘ignore’ - the unencodable character is ignored
#   ‘replace’ - the unencodable character is replaced by questionmark

# Buffering is storing the information from the input(writing to file) into a buffer and not directly sending it to the file
# Therefore, if 1 is specified as the buffer, then input is written to file once a newline is reached. If a buffer size of
# 500 is specified, when the buffer is full, input is written to file. If 0 is specified, then buffering is off and files
# are written directly to the file. By default, writing to the file happens when file object is closed(removed from
# temporary memory).

# It is useful when using large file sizes with high latency (i.e. input from source is slow, then with buffering the 
# program has to make fewer system calls to write and program speed is faster). On the other hand, working with large file
# sizes with fast streaming data may overflow the buffer causing a system issue (like a crash).

# The effects of buffering will not be visible until we have huge file sizes and/or latency - therefore, we shall skip 
# examples of buffering parameter.

#Encoding and errors - parameters are the same as for bytes objects.
```

```python
with open('animal_jokes.txt', encoding = 'ascii', errors = 'backslashreplace')as f:
    for line in f.readlines():
        print(line)
```

```python
#Method          Description
# close()        Closes the file  - X
# detach()       Returns the separated raw stream from the buffer - X
# fileno()       Returns a number that represents the stream, from the operating system's perspective - X
# flush()        Flushes the internal buffer - X
# isatty()       Returns whether the file stream is interactive or not - X
# read()         Returns the file content - X
# readable()     Returns whether the file stream can be read or not - X
# readline()     Returns one line from the file - X
# readlines()    Returns a list of lines from the file - X
# seek()         Change the file handle position - X
# seekable()     Returns whether the file allows us to change the handle position - X
# tell()         Returns the current file handle position - X
# truncate()     Resizes the file to a specified size - X
# writable()     Returns whether the file can be written to or not - X
# write()        Writes the specified string to the file - X
# writelines()   Writes a list of strings to the file - X
```

### Binary files. Bytes and bytesarray objects

```python
bytes1 = bytes([145,189,210,255])

print(bytes1)
```

```python
#bytes object will only convert values between 0 to 255 into hexadecimal. Below 0 and above 255 will throw an error

bytes1 = bytes([256])
```

```python
# If the value that we want to convert to bytes - is a printable ASCII character, then though in the back end the data is 
# stored as hexadecimal, on our terminals the corresponding ascii character will be displayed. 

bytesAscii = bytes([5,33,65,97])
print(bytesAscii)
```

```python
# The bytes function takes one optional parameter. 
# If no parameters provided - then creates an empty bytes object.

bytes1 = bytes()

print(bytes1)
print(type(bytes1))
```

```python
# Dealing with binary files

with open('WaterMelon32x32.jpg', 'rb') as img_f:
    data = img_f.readlines()
    for x in range(len(data)):
        print(f'Line no. {x}, length : {len(data[x])} and type : {type(data[x])}.')
        print('-'*125)
        print(f'Data : {data[x]}.')
    
    
```

```python
with open('WaterMelon32x32.jpg', 'rb') as img_f, open('WaterMelon32x32Rev1.jpg', 'wb') as img_W:
    data = img_f.readlines()
    data[16] = bytes([145,179,210,199])
    print(data[16])
    img_W.writelines(data)
```

```python
with open('WaterMelon32x32.jpg', 'rb') as img_f, open('WaterMelon32x32Rev2.jpg', 'wb') as img_W:
    data = img_f.readlines()
    data12 = bytearray(data[12])
    bytesArrObj = []
    for elem in range(1,11):
        bytesArrObj.append((data[12][elem]+100)%255)
    data12[1:11] = bytearray(bytesArrObj)
    print(data12)
    print(type(data12))
    data[12] = bytes(data12)
    img_W.writelines(data)
```
