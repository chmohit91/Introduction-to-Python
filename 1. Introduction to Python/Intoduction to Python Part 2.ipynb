# - **High Level langauge**

- It is a human friendly computer lanaguage
- The code is written in simple langauge such as english
- It is easier to read and understand for a human being
- Example - C, C++, Java, Python, Ruby, R etc

# - **Low level langauge**

- It is a machine friendly computer langauge
- The code written in this format is in binary digit also known as bits (0,1)
- It is almost impossible for us human to understand it
- Example - assembly language, Pascal , fortrain etc


# **Translator**

- It is a software that converts the high level language into a machine langauge and vice versa

## **Types of Translator**

- Compiler
- Interpreter


### **Compiler**

- A compiler is a software that reads the whole high level code at once and it converts the whole code in machin langauge at once


### **Interpreter**

- An interpreter is a software that reads the high level code line by line and will convert it into machine language line by line
- It is going to read and convert one line at a time

# **Compiled Language and Interpreted langauge**

## **Compiled langauge**

- Those high level langauges that uses compiler as their translator are known as compiled langauges
- Example - C, C++, Java, etc

## **Interpreted Language**

- Those high level langauges that uses interpreter as their translator are known as interpreted langauges
-Example - Python, R, Ruby, PHP etc

# **Python - Both a compiled langauge and interpreted language**

Python is both compiled and interpreted language, it has two main steps involved in it:-

- Compilation --> When we write a code it will be first compiled into a lower level language called as a bytecode, this compilation is done with the help of a compiler

- Execution --> The bytecode is then converted into machine language by the interpreter line by line




# **Source code:-**

- The code that we write as a developer is known as source code
- the extension of the python sourece code is .py

# **Bytecode or Intermedite code:-**

- As soon as the source code is executed (run), compilation of the source code starts
- This whole step is known as compilation because at the end we are going to have one code written in a langauge that still needs a translator to convert it into a machine langauge.
- Here the code will be converted in one go
- The extension for your bytecode is .pyc
- IF While compiling the compiler finds any error, bytecode generation will not going to happend

# **Machine code**

- As soon as we get the bytecode, the interpreter comes into play
- the interpreter converts the whole bytecode into machine code line by line
- As soon as the error is detected, the execution will stop
- in this case your code will be partially executed
- If no error is found the code will be converted into the machine code
- As soon as the bytecode is converted to machine code, the bytecode gets deleted or destroyed
